```python
import re
import requests
from bs4 import BeautifulSoup as bs

def removePunctuation(source):
    xx = u"([^\u4e00-\u9fff]+)"
    s = re.sub(xx,' ',source)
    return s


def removeTag(soup,tag):
    [x.extract() for x in soup.select(tag)]


res = requests.get('http://blog.cw.com.tw/blog/profile/260/article/1826')
soup = bs(res.text,'html.parser')
removeTag(soup, 'strong')
words = soup.select('p')[4:14]
word = ''
for w in words:
    word += w.text
cleanw =  removePunctuation(word)
wordlist = cleanw.split()
with open('./Workspace/data/dict.txt', 'w') as f:
    for w in wordlist:
        f.write(w.encode('utf-8') + '\n')
```

```python
import requests
import json
from bs4 import BeautifulSoup as bs

home = requests.get('https://price.nat.gov.tw/p/zh_tw/index', verify = False)
soup = bs(home.text)
cat = soup.select('.price_tag.img-circle')



head = {
    'User-Agent':'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_11_3) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/50.0.2661.102 Safari/537.36'
}
payload = {
    'class_uuid':'',
    'page':'1'
}

for item in cat:
    payload = {
        'class_uuid':item['data-info'],
        'page':'1'
    }
    res = requests.post('https://price.nat.gov.tw/p/zh_tw/price_center_list/all',data=payload, headers = head, verify=False)
    jd = json.loads(res.text, encoding = 'utf8')
    for ele in jd['material']:
        print jd['material'][ele]['name'], jd['material'][ele]['min_price'], jd['material'][ele]['max_price']
    while jd['page']['NEXTPAGE'] > jd['page']['NOWPAGE']:
        payload['page'] = jd['page']['NEXTPAGE']
        res = requests.post('https://price.nat.gov.tw/p/zh_tw/price_center_list/all',data=payload, headers = head, verify=False)
        jd = json.loads(res.text, encoding = 'utf8')
        for ele in jd['material']:
            print jd['material'][ele]['name'], jd['material'][ele]['min_price'], jd['material'][ele]['max_price']
```

```python
import grequests
from bs4 import BeautifulSoup as bs
import time
start = time.time()
textSet = set()

url = 'http://bookmarks.tw/dictionary/archives/category/uncategorized/page/{}'
header = {
    'User-Agent':'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_11_3) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/50.0.2661.102 Safari/537.36'
}

rs = (grequests.get(url.format(p), headers = header) for p in xrange(1,5294))
resp = grequests.map(rs, size = 20)

count = 0
for res in resp:
    soup = bs(res.text)
    for t in soup.select('.entry-title'):
        text = [ele for ele in t.text.split(u'…') if ele != '' and len(ele) >= 2]
        for tx in text:
            textSet.add(tx)
```

```python
with open('./Workspace/dict_1.txt', 'w') as f:
    for i in list(textSet):
        f.write(i.encode('utf-8')+'\n')
```
