```python
class tree():
    def __init__(self):
        self.trie = {}
    def tree(self, word):
        p = self.trie
        for c in word[0]:
            if c not in p:
                p[c] = {}
            p = p[c]
        p[''] = word[1]
    def search(self, word):
        p = self.trie
        for c in word:
            try:
                p = p[c]
            except KeyError:
                return None
        output = p.values()[0]
        return output
```

```python
import random

def randomAlpha():
    alphabeta = ['a','b','c','d','e']
    out = [alphabeta[random.randint(0,4)] for ele in alphabeta]
    return ''.join(out)
```

```python
treeobj = tree()
number = [str(ele) for ele in xrange(100000,150000)]
tag = ['%s%d' %(randomAlpha(),ele) for ele in xrange(100000,150000)]
data = zip(number ,tag)
for n in data:
    treeobj.tree(n)
#print treeobj.trie
```
